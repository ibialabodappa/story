# DATASET STATISTICS

## The list of house prices( in thousands of dollars) 

```python
prices=[250,320,180,450,290,310,275,390,210,340]
```

### Total sum of all the house prices
using for loop to add each number in the list and then add each of them every time the loop runs

```python
total=0
for i in prices:
  total+=i
print(f"the sum of prices for the houses are ${total}")
```

### Mean average price of the houses

 incrimenting or adding one to the defined function to count how meny element are in the list since len() is prohibited

```python
count=0
for i in prices:
    count+=1
print(count)
```

calculating the average by dividing total which is the sum of the houses by the number of element in the list which is count


```python
average=total/count
print(f"The average price of the house is ${average}")
```

### checking the minimum and maximum prices of the house
checking the maximun and maximum value i have to create an empty list for each and use a for loop to loop through the prices list and also created a variabel maximal which is equal to zero
1. after that i use if condition to check if each element inside the loop is greather than maximal
2. if yes maximal will be equal to n and i appended it to max_list which is used to store highest value when checked
3. if no i used else to append the small number to the mini_list when checked 
4. then i equated the last number which is the highest using index method -1 for maximum price, the same for minimum but i used index 0 since we are looking for the smallest number 
5. finally i printed it out using f string

```python
maximal=0
mini_list=[]
max_list=[]
for n in prices:
    if n>maximal:
      maximal=n
      max_list.append(maximal)
    else:
        mini_list.append(n)
maximum_price=max_list[-1]
minimum_price=mini_list[0]
print(f"The highest house price is ${maximum_price}\nThe lest price for the houses is ${minimum_price}")
```

### the range of (max-min)
subtracting the maximum price from the minimun price in the list of house prices

```python
list_range=maximum_price-minimum_price
print(f"The range between maximum and minimum price of the house {list_range}")
```

### houses that cost more than $300,000 
i used for loop too loop through and then checked each price of house that is higher than 300 after that i added 1 to the summation variable created earlier.after all i printed the summation using f string

```python
sumation=0
for i in prices:
    if i>300:
        sumation+=1
print(f"The number of house that cost more than $300,000 is {sumation}")
```

# CLASSIFICATION METRICS
firstly initallizing variables

```python
True_positive=45
True_negative=32
False_positive=8
False_negative=15
```

#### calculating the accuracy,precision,recall,f1 score use arithemetic operations

```python
accuracy=(True_positive+True_negative)/(True_positive+True_negative+False_positive+False_negative)
print(f"Accuracy {accuracy}")
```

```python
precision=(True_positive)/(True_positive+False_positive)
print(f"Precision {round(precision,4)}")
```

```python
recall=(True_positive)/(True_positive+False_negative)
print(f"Recall {recall}")
```

```python
fi_score=2*(precision*recall)/(precision+recall)
print(f"F1 score {round(fi_score,4)}")
```

# FEATURE  SCALING

2. normalizing each features using Min-Max scalling to range of [0,1],given
3. Calculating each normilized valuse to know the highest


```python
feature_A=85
feature_B=1200
feature_C=0.45
```

```python
#Maximum values for each features
A_max=100
B_max=2000
C_max=0.9
```
